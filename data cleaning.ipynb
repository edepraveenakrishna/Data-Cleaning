# Checking null values

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace = True)
df.shape
```

# Cleaning rate column

```python
df['rate'].unique()
```

```python
def handlerate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)
    
df['rate'] = df['rate'].apply(handlerate)
df['rate'].head()
```

```python
df['rate'].fillna(df['rate'].mean(), inplace = True)
df['rate'].isnull().sum()
```

```python
df.dropna(inplace = True)
df.head()
```

```python
df.rename(columns = {'approx_cost(for two people)':'cost for 2 people', 'listed_in(type)':'Type', 'listed_in(city)':'city'}, inplace = True)
df.head()
```

# Cleaning location column

```python
df['location'].unique()
```

```python
df['city'].unique()
```

```python
df = df.drop(['city'], axis = 1)
df.head()
```

# Cleaning cost for 2 people column

```python
df['cost for 2 people'].unique()
```

```python
def handlecomma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',', '')
        return float(value)
    else:
        return float(value)
    
df['cost for 2 people'] = df['cost for 2 people'].apply(handlecomma)
df['cost for 2 people'].unique()
```

```python
df.head()
```

# Understanding highest and lowest values in various columns

```python
rest_types = df['rest_type'].value_counts(ascending  = False)
rest_types
```

```python
rest_types_lessthan1000 = rest_types[rest_types<1000]
rest_types_lessthan1000
```

```python
location = df['location'].value_counts(ascending  = False)
location_lessthan300 = location[location<300]
location_lessthan300
```

```python
def handle_location(value):
    if(value in location_lessthan300):
        return 'others'
    else:
        return value
        
df['location'] = df['location'].apply(handle_location)
df['location'].value_counts()
```

```python
cuisines = df['cuisines'].value_counts(ascending  = False)
cuisines_lessthan100 = cuisines[cuisines<100]
cuisines_lessthan100
```

```python
def handle_cuisines(value):
    if(value in cuisines_lessthan100):
        return 'others'
    else:
        return value
        
df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```
